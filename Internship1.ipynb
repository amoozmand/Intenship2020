```python
from sentinelsat import SentinelAPI, read_geojson, geojson_to_wkt
from datetime import date
```

```python
# connect to the API

api = SentinelAPI('****','*********','https://scihub.copernicus.eu/dhus')
```

```python

footprint='POLYGON ((4.4245265 50.8119984, 4.4245142 50.812042, 4.424705 50.8120635, 4.424709 50.8120492, 4.424738 50.8119465, 4.4245472 50.811925, 4.4245265 50.8119984))'
products = api.query(footprint,
                 date=('20180901', date(2019, 9, 5)),
                 platformname='Sentinel-2',
                cloudcoverpercentage=(0, 10))
```

```python
products_gdf = api.to_geodataframe(products)
```

```python
products_gdf_sorted = products_gdf.sort_values(['cloudcoverpercentage'], ascending=[True])
```
